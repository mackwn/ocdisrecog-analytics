```python
def extract_img_features(fn, imgdir='data/preprocessed_images'):
    '''
    Take input filename for preprocessed fundus image.
    Assumes resolution of 512 px by 512 px. 
    Returns a 1 dimensional row vector
    '''
    path = os.path.join(imgdir, fn)
    img = cv2.imread(path)
    try:
        img = img.reshape(1, 512*512*3)
    except:
        raise ValueError('Unable to reshape image. Ensure the image is 512 px by 512 px')
    
    return img
    
def df_to_imagesdf(df, imgdir='data/preprocessed_images'):
    '''
    Take input dataframe and add columns for pixel values for that rows images. Returns a new dataframe with 
    the existing data and the new image data.
    
    Keyword Arguments:
    df -- dataframe of image data. Assumes df has a column for image filename. Images must confirm to requirements
        for extract_img_features function
    imgdir -- location of images refrences in dataframe filename column
    '''
    
    # Extract data from dataframe
    data = df.values
    data_cols = list(df.columns.values)
    # Make sure filename is passed
    if 'filename' not in data_cols: raise ValueError('Dataframe must contain a filename column.')
    
    # Iterate over each row of data an extract an image for each filename
    img_features = None
    for row in data:
        fn = row[df.columns.values=='filename'][0]
        # print(fn)
        img = extract_img_features(fn, imgdir)
        if img_features is None: img_features=img
        else: img_features=np.concatenate((img_features, img))

    img_features = np.array(img_features)  
    
    #Concatenate data into a new dataframe
    pxcol_names = ['Pxcol_{}'.format(val) for val in range(0, img_features.shape[1])]
    
    image_df = pd.DataFrame(img_features, columns=pxcol_names)
    data_df = pd.DataFrame(data, columns=data_cols)
    
    return pd.concat((data_df, image_df), axis=1)
    
    #return pd.DataFrame(np.concatenate((data, img_features), axis=1), columns=data_cols+pxcol_names)


def df_to_imagesdf_array(df, imgdir='data/preprocessed_images'):
    '''
    Take input dataframe and add a single column storing an array of image data. Returns a new dataframe with 
    the existing data and the new image data.
    
    Keyword Arguments:
    df -- dataframe of image data. Assumes df has a column for image filename. Images must confirm to requirements
        for extract_img_features function
    imgdir -- location of images refrences in dataframe filename column
    '''
    
    # Extract data from dataframe
    data = df.values
    data_cols = list(df.columns.values)
    # Make sure filename is passed
    if 'filename' not in data_cols: raise ValueError('Dataframe must contain a filename column.')
    
    # Iterate over each row of data an extract an image for each filename
    img_features = None
    for row in data:
        fn = row[-1]
        img = extract_img_features(fn, imgdir)
        if img_features is None: img_features=img
        else: img_features=np.concatenate((img_features, img))

    img_features = np.array(img_features)  
    
    #Concatenate data into a new dataframe
    pxcol_names = ['Pxcol_{}'.format(val) for val in range(0, img_features.shape[1])]
    return pd.DataFrame(np.concatenate((data, img_features), axis=1), columns=data_cols+pxcol_names)

def plot_image_grid(df, col_wrap=5):
    pass
```

## Image Properties

### Avg Pixel Values

```python
def extract_avg_colors(fn, dir='data/preprocessed_images'):
    '''
    Take input filename for preprocessed fundus image.
    Assumes resolution of 512 px by 512 px. 
    Returns a 1 dimensional row vector
    
    Keyword Arguments:
    fn -- filename for fundus image
    dir -- location of fundus image files
    
    '''
    path = os.path.join(dir, fn)
    img = cv2.imread(path)
    try:
        avg_colors = img.mean(axis=(0,1))
    except:
        raise ValueError('Unable to reshape image. Ensure the image is 512 px by 512 px')
    
    return avg_colors.reshape(1,-1)

data = df[['ID', 'Patient Age', 'Patient Sex', 'N', 'D',
       'G', 'C', 'A', 'H', 'M', 'O', 'filename']].values

def df_to_avgcolordf(indf):
    '''
    Take input dataframe of fundus images and returns a dataframe 
    with columns for the average value for each color for each image.
    
    Keyword Arguments:
    df -- dataframe containing ocular disease classifcation data, must include columns 'ID',
        'Patient Age', 'Patient Sex', 'N', 'D','G', 'C', 'A', 'H', 'M', 'O', 'filename'
    '''
    
    data = indf[['ID', 'Patient Age', 'Patient Sex', 'N', 'D',
       'G', 'C', 'A', 'H', 'M', 'O', 'filename']].values
    color_vals=None
    for row in data:
        fn = row[-1]
        colors = extract_avg_colors(fn)
        #print(colors.shape)
        #if color_vals is not None: print(color_vals.shape)
        if color_vals is None: color_vals=colors
        else: color_vals=np.concatenate((color_vals, colors))
    
    ## concatenate data frames and not arrays - otherwise fucks up the dtypes
    return pd.DataFrame(np.concatenate((data, color_vals), axis=1),
             columns = ['ID', 'Patient Age', 'Patient Sex', 'N', 'D',
       'G', 'C', 'A', 'H', 'M', 'O', 'filename', 'Red_mean', 'Green_mean', 'Blue_mean'])
```

```python
color_df = df_to_avgcolordf(df)
```

```python
def diagnosis_identifier(row):
    diagid = ''
    #print(row)
    for key in ['N', 'D','G', 'C', 'A', 'H', 'M', 'O']:
        if (int(row[key]) == 1):
            #print('yes')
            #print(key)
            diagid+=key
            #print(key)
    if len(diagid)==0: diagid='NAN'
    return diagid
```

```python
img_df2['diag_id'] = img_df2.apply(lambda row: diagnosis_identifier(row), axis=1)
```

```python
mean_images_df = img_df2.loc[:, ['diag_id']+pxcol_names].groupby('diag_id').mean()
```

```python
def plot_image_grid(images, labels, config={'figsize':(12,12), 'pad':0.5, 'cols':4}):
    
    #images = imgdf.values
    fig = plt.figure(figsize=config['figsize'])
    cols = config['cols']
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     nrows_ncols=(math.ceil(mean_images.shape[0]/cols), cols),  # creates 2x2 grid of axes
                     axes_pad=config['pad'],  # pad between axes in inch.
                     )

    i=0
    for ax, im in zip(grid, images):
        # Iterating over the grid returns the Axes.
        ax.imshow(im.reshape((512,512,3))/255)

        label = labels[i]
        i+=1
        ax.set_title('{}'.format(label))



        ax.grid(False)
    #plt.tight_layout()
    plt.show()

plot_image_grid(mean_images_df.values, mean_images_df.index.values)
```
